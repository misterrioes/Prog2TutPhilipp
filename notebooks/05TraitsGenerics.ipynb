## Vererbung: Traits <- konkrete klasse
### Traits
- ähnlich zu Interfaces dürfen aber auch Felder haben.
- unterstützen Mehrfachvererbung        
- **können nicht instanziiert** werden.
- Traits können abstrakte und nicht-abstrakte Methoden enthalten.
- **haben keine Konstruktorparameter** (anders als abstrakte Klassen).
- werden mit `extends` (erste Vererbung) und `with` (weitere Traits) kombiniert.

````scala

trait Wirtschaft {
  val arbeitsPensum = "hoch"  
  def learnWirtschaft(): String ="Ich lerne Wirtschaft"
}

trait Informatik {
  def learnInformatik(): String = "Ich lerne Informatik"

}
abstract class Student {
  def study(): String
}
class InformatikStudent extends Student with Informatik   {
  override def study(): String = "Ich studiere reine Informatik"
}



val einStudent = new InformatikStudent with Wirtschaft
println(einStudent.study())
println(einStudent.learnWirtschaft())
println(einStudent.learnInformatik())
````

## Dynamic Binding
- Dynamic Binding bezeichnet die Technik, bei der bei überschriebenen Methoden erst zur Laufzeit anhand des tatsächlichen Objekttyps entschieden wird, welche Methodenimplementierung ausgeführt wird – nicht anhand des statischen Typs der Referenz.
- gilt nur für überschriebene Methoden mit override. 
- Die Auswahl der konkreten Methode erfolgt zur Laufzeit, nicht beim Kompilieren

````scala
class Reptilie(val legtEier:Boolean){
    def schwimmen(): String= "Vielleicht kann ich Schwimmen vielleich nicht"
}
class Schildkröte(val panzerung:Double) extends Reptilie(true){
    override def schwimmen():String = "Ich kann Schwimmen"
}

val aReptile: Reptilie = new Schildkröte(50) 
aReptile.schwimmen()// hier wird die Schildkrötenmethode ausgeführt, obwohl wir eigentlich ein Reptilien Variable haben



````

## Typschranken
- werden eingesetzt, um Typparameter auf bestimmte Typbereiche einzuschränken
- besonders nützlich bei Vererbungshierarchien:

![image](./images/05/Klassendiagramm.png)

````scala
class Zoo[A](val theAnimals: Array[A])
// Zoo von Fischen
val aquarium:Zoo[Fish] = new Zoo(new Array[Fish](0))
//=> `A` ist der Typparameter. Beim Erstellen einer Instanz wird der konkrete Typ festgelegt:
// Zoo von Menschen
val humarium: Zoo[Human] = new Zoo(Array[Human]())
// Zoo mit Fischen bewohnt
val aquariumBewohnt: Zoo[Fish] = new Zoo(Array(new Fish, new Fish))
````

````scala

class Being{
    def call():String = "Ich bin ein Being"
}
class Animal extends Being 
class Human extends Being{
 override def call():String = "Ich bin ein Mensch hol mich hier raus"
}
class Vertebrates extends Animal
class Invertebrates extends Animal
class Mammals extends Vertebrates
class Birds extends Vertebrates
class Fish extends Vertebrates
class Insects extends Invertebrates
class Arachnids extends Invertebrates
class Wildlife extends Mammals
class Pet extends Wildlife {  
    override def call(): String = "Ich bin ein Pet. Streichel mich"
}
class Dog extends Pet 
class Cat extends Pet {
   override def call():String=  "Ich bin eine Katze. Fütter mich."
}
val a1:Being = Human()
val a2: Being = Dog()
val a3: Being = Cat()
println(a1.call())
println(a2.call())
println(a3.call())
````

### Obere Typschranke A <: SomeClass
- legt fest dass A eine Unterklasse von SomeClass sein muss

````scala

class Zoo[A <: Animal](animals: Array[A])
// *A* muss vom Typ `Animal` oder einer Unterklasse davon sein.
new Zoo[Animal](Array[Animal]())     // erlaubt
new Zoo[Pet](Array[Pet]())           // erlaubt
//new Zoo[Human](Array[Human]())       // nicht erlaubt

````

#### Untere Typschranke (`>:`)


````scala
class Zoo[A >: Wildlife](theAnimals: Array[A])

//=> *A* muss vom Typ `Wildlife` oder einer Oberklasse davon sein.

new Zoo[Mammals](Array[Mammals]())     // erlaubt
new Zoo[Wildlife](Array[Wildlife]()) // erlaubt
//new Zoo[Birds](Array[Birds]())// nicht erlaubt
````

### 1. Kovarianz (`+T`)
Die Subtyp-Beziehung wird **übertragen**: 
Wenn `A <: B`, dann gilt auch `Container[A] <: Container[B]`.
```scala
// Wenn Dog <: Animal, dann ist auch Container[Dog] <: Container[Animal]
class Container[+T](val content: List[T])
//Main:
val dogContainer: Container[Dog] = new Container[Dog](List(new Dog()))
val animalCage: Container[Animal] = dogContainer  // funktioniert wegen +T => siehe Subtyp-Polymoprhie, falls Verständnisproblem
```

Hier werden keine Klassenparameter als Inputtypen der Methoden der generischen Klassen erlaubt also:
```scala 
class  Klasse [+T]
    def function (input T): outputTyp 
```
wäre nicht erlaubt.
Der Grund liegt darin dass wir evtl eine Variable von Obertypen auf den Subtypen setten könnten in dieser Methode und das würde einen Kompilierfehler entstehen lassen.

### 2. Kontravarianz (`-T`)
Die Subtyp-Beziehung wird **umgekehrt** übertragen:
Wenn `A <: B`, dann gilt `Container[B] <: Container[A]`.
```scala
// Wenn Dog <: Animal, dann gilt Trainer[Animal] <: Trainer[Dog]
class Trainer[-T] {
  def train(tier: T): Unit = println("Trainiert ein Tier")
}
val animalTrainer: Trainer[Animal] = new Trainer[Animal]
val dogTrainer: Trainer[Dog] = animalTrainer  // erlaubt wegen -T: Ein Trainer[Animal] kann überall eingesetzt werden, wo ein Trainer[Dog] erwartet wird
```
Hier sind nur produzierenden Methoden erlaubt, also Methoden welche einen output vom Typ T zurückgeben wären nicht erlaubt. Das bedeutet 
```scala
class Klasse [-T]
    def function(input irgendeinTyp) :output T 
```
wäre nicht erlaubt.
Der Grund liegt darin dass wir evtl eine Variable von Obertypen auf den Subtypen setten könnten wenn wir ein Animal bekommen obwohl wir einen Dog erwarten aus der Gettermethode und  das würde einen Kompilierfehler entstehen lassen.

### 3. Invarianz (`T`)
Es findet **keine** Übertragung der Subtyp-Beziehung statt:
`Container[A]` und `Container[B]` sind **unabhängig**, auch wenn `A <: B` gilt.
Ist der **Default**, wenn keine Varianz-Angabe gemacht wird
```scala
// Keine Subtyp-Beziehung: Box[Dog] und Box[Animal] sind inkompatibel, auch wenn Dog <: Animal
class Box[T](var inhalt: T)
val dogBox: Box[Dog] = new Box(new Dog)
// val animalBox: Box[Animal] = dogBox  // nicht erlaubt
```
