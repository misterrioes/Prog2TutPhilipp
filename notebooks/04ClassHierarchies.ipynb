## Verberung: konkrete/abstrakte Klasse <- konkrete Klasse

- **Subklasse**: die erbende Klasse

### abstrakte Klassen
- kГ¶nnen **nicht instanziiert** werden.
- kГ¶nnen abstrakte Methoden und abstrakte Felder enthalten, d.h. Member ohne Implementierung.
- eine Klasse, die von einer abstrakten Klasse erbt, **muss** alle abstrakten Member (Гјberschreiben).
- bereits implementierte Member in der abstrakten Klasse **kГ¶nnen** Гјberschrieben werden, **mГјssen aber nicht**.
- eine Klasse die eine abstrakte Methode oder Feld hat muss auch abstrakt gekennzeichnet sein sonst gibt es einen Compilerfehler.
