### Option[T]
`Option[T]` ist ein abstrakter(!) Typparameter mit nur zwei möglichen Ausprägungen:
1. `Some(T)` 
2. `None`


```scala
val x: Option[Int] = Some(5)
val y: Option[Int]= None
println(x) //5
println(y)//None


///Fehlerbehandlungen mit Option[T]

def divide(a:Int, b:Int): Int = a/b
//println(divide(3,0)) //Wirft eine Arithmetic exception!

def divideAberSafe(a: Int, b: Int): Option[Int] = {
  if (b == 0) None else Some(a / b)
}
println(divideAberSafe(3,0)) //printed None
```

# Java Streams
- **Definition**: Sequenz von Elementen, die von einer
Quelle erzeugt und auf denen Operationen zur
Datenverarbeitung angewendet werden
- ermöglichen deklarative Programmierung, die sich auf die Geschäftslogik funtkoniert
Stream-*Pipeline:*
![image](./images/07/StreamUsageOverview.png)
1. `.stream()` erzeugt eine Stream
2. `intermediate Operations` verarbeiten die Elemente der Pipeline nach Belieben (`Lambda-Ausdrücke` oder `Methodenreferenzen` bestimmen das Verhalten der i.o.)
3. Erst mit `einer terminalen Operation` wird die Pipeline ausgeführt und ein Ergebnis zurückgeliefert

![image](./images/07/StreamIntermediateTerminal.png)


## Intermediate Operations:
- arbeiten auf Stream, **Geben neuen Stream zurück**
- sind "lazy" => sie werden erst ausgeführt, wenn eine `terminale Operation` folgt
- werden unterteilt in "Stateless" (zustandslose) und "Stateful" (zustandsbehfatete) (stateless können sorgenfrei parallel ausgeführt werden):
    - `Stateless`
Die Operation wird unabhängig von anderen Elementen im Stream ausgeführt. Sie hat kein Gedächnis.
    - `Stateful`
Die Operation benötigt Kontextinformationen über andere Elemente (z. B. vorherige oder spätere). Sie hat ein Gedächnis!
### Übersicht Intermediate Operations: Stateless
| Methode     | Functional Interface       | Lambda-Signatur        | Beispiel                         | Rückgabetyp     |
|-------------|----------------------------|------------------------|----------------------------------|------------------|
| `filter`    | `Predicate<T>`             | `T → boolean`          | `n -> n % 2 == 0`                | `Stream<T>`     |
| `map`       | `Function<T, R>`           | `T → R`                | `s -> s.length()`                | `Stream<R>`     |
| `flatMap`   | `Function<T, Stream<R>>`   | `T → Stream<R>`        | `s -> Stream.of(s.split(""))`   | `Stream<R>`     |
| `mapToInt`  | `ToIntFunction<T>`         | `T → int`              | `s -> s.length()`                | `IntStream`     |
| `mapToDouble`| `ToDoubleFunction<T>`     | `T → double`           | `s -> Math.sqrt(s)`              | `DoubleStream`  |
| `mapToLong` | `ToLongFunction<T>`        | `T → long`             | `s -> s.length() * 1L`           | `LongStream`    |

## Terminale Operationen:
- arbeiten auf Stream und geben ein konkretes Ergebnis zurück (z. B. `List`, `Optional`, `long`, `boolean`, etc.) => beenden Streampipeline

### Übersicht Terminale Operationen
| Methode     | Functional Interface       | Lambda-Signatur        | Beispiel                             | Rückgabetyp       |
|-------------|----------------------------|------------------------|--------------------------------------|--------------------|
| `collect`   | `Collector<T, A, R>`       | –                      | `Collectors.toList()`               | `List<T>`, `Map`, etc. |
| `reduce`    | `BinaryOperator<T>`        | `(a, b) → T`           | `(a, b) -> a + b`                    | `Optional<T>` / `T` |
| `forEach`   | `Consumer<T>`              | `T → void`             | `System.out::println`               | `void`             |
| `min`       | `Comparator<T>`            | `(a, b) → int`         | `Comparator.comparing(...)`         | `Optional<T>`      |
| `max`       | `Comparator<T>`            | `(a, b) → int`         | `Comparator.comparing(...)`         | `Optional<T>`      |
| `count`     | –                          | –                      | –                                    | `long`             |
| `anyMatch`  | `Predicate<T>`             | `T → boolean`          | `x -> x > 5`                         | `boolean`          |
| `allMatch`  | `Predicate<T>`             | `T → boolean`          | `x -> x > 0`                         | `boolean`          |
| `noneMatch` | `Predicate<T>`             | `T → boolean`          | `x -> x < 0`                         | `boolean`          |
| `findFirst` | –                          | –                      | –                                    | `Optional<T>`      |
| `findAny`   | –                          | –                      | –                                    | `Optional<T>`      |
| `sum`       | –                          | –                      | `.mapToInt(...).sum()`              | `int` / `double` / `long` |
| `average`   | –                          | –                      | `.mapToInt(...).average()`          | `OptionalDouble`   |
#### Collectors-Objekt
- wird nur in Kombi. mit der terminalen Operation `.collect(...)` verwendet
- enthält eine Sammlung an Operationen, mit denen man den Rückgabetyp der Pipeline festlegt

| Methode           | Zweck                                      | Beispiel                         | Rückgabetyp         |
|-------------------|--------------------------------------------|----------------------------------|----------------------|
| `toList()`        | Stream → Liste                             | `collect(Collectors.toList())`  | `List<T>`            |
| `toSet()`         | Stream → Menge                             | `collect(Collectors.toSet())`   | `Set<T>`             |
| `toMap()`         | Stream → Map                               | `collect(Collectors.toMap(...))`| `Map<K, V>`          |
| `groupingBy()`    | Gruppiert nach einem Kriterium             | `groupingBy(Student::getUni)`   | `Map<K, List<T>>`    |
| `counting()`      | Zählt Elemente                             | `groupingBy(..., counting())`   | `Long`               |
| `joining()`       | Strings zusammenfügen                      | `joining(", ")`                 | `String`             |

## Methodenreferenz
- Alternative zu Lambda-Ausdrücken => bestimmt das Verhalten von intermediären oder terminalen Operationen
- Verhalten = Verweis auf existierende Methode
- Syntax: **Zielreferenz :: Methodenname**

| Nr. | Art | Beispiel | Entspricht (Lambda) | Erklärung | Wie wird das Stream-Element verwendet? |
|-----|------|----------|----------------------|------------|------------------------------------------|
| 1   | **Statische Methode einer Klasse** | `Integer::sum`, `Math::max` | `x -> Integer.sum(x, ...)`, `x -> Math.max(x, ...)` | Referenz auf eine `static`-Methode | Stream-Element(e) werden **als normale Parameter übergeben** |
| 2   | **Instanzmethode eines Typs** | `String::toUpperCase` | `s -> s.toUpperCase()` | Methode wird auf jedes Element angewendet | Stream-Element wird **zum `this`-Objekt (receiver)** |
| 3   | **Instanzmethode eines konkreten Objekts** | `System.out::println` | `s -> System.out.println(s)` | Methode eines festen Objekts wird auf jedes Element angewendet | Stream-Element wird **als Parameter an dieses feste Objekt übergeben** |
| 4   | **Konstruktorreferenz** | `Student::new` | `x -> new Student(x)` | Erzeugt neue Objekte mit Konstruktor | Stream-Element(e) werden **als Konstruktorparameter verwendet** |
