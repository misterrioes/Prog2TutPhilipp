## For-Schleifen als Ausdruck
- kann Collections durchlaufen und transformieren
- mit `yield`: erzeugt eine neue Collection mit transformierten Werten
- der Collection-Typ der erstgenutzen Collection bleibt erhalten, der Elementtyp kann sich ändern (z. B. `List[Int]` → `List[String]`)
- Guards sind Bedingungen an welche tatsächlich genutzt werden 

```java
List<String> listOfNameAndAge = new ArrayList<String>()
for (String name: listOfNames) {
    if( name.length >3){
    for( int age: new Array(20,30)){
       listOfNameAndAge.add(name + "(" + age + ")")
    }
}
}

````scala
val listOfNames:List[String] = List("Anna", "Bob", "Chris")
val result = for {
  name <- listOfNames
  if name.length > 3
  age <- List(20, 30)
} yield s"$name ($age)"

println(result)



val resultInternalCalculations = listOfNames
  .filter(name => name.length > 3)
  .flatMap(name =>
    List(20, 30)
      .map(age => s"$name ($age)")
  )

println(resultInternalCalculations == result)

````

# Nebenläufigkeit

## Definition: Prozess, Thread

Ein **Prozess** ist die Ausführung eines Programms: 
![image](./images/09/ProcessThread.png)


Ein **Thread** ist die kleinste Ausführungseinheit innerhalb eines Prozesses:

- Jeder Prozess hat **mindestens einen** (Main-Thread)
- Threads **teilen sich**:
  - Code
  - Heap
  - Ressourcen
- Sie **besitzen jeweils einen eigenen Stack**

### Speicherbereiche: Heap vs. Stack
- **Heap:**  Speicherbereich, den alle Threads eines Prozesses gemeinsam nutzen  
  Enthält:
  - Objekte
  - Instanzvariablen
  - Klassen- bzw. Singleton-Variablen

- **Stack:**  Jeder Thread hat seinen eigenen Stack   
  Enthält:
  - Lokale Variablen (in Java)
  - Methodenparameter
  - Referenzen auf Heap-Objekte
  - funktioniert nach LIFO

```scala
object Konfiguration {                  // Singleton → Heap
  var sprache: String = "Deutsch"      // Klassenvariable → Heap
}

class Benutzer(val name: String) {     // Instanzvariable → Heap
  var status: String = "offline"       // Instanzvariable → Heap

  def einloggen(passwort: String): Unit = {
    val msg = "Willkommen!"            // Lokale Referenzvariable -> Referenz im Stack Wert im Heap
    status = "online"
    println(msg + " " + name)
  }
}
```
#### Wo wird was gespeichert?

- Instanzvariablen → Heap
- Klassenvariablen / Singleton-Variablen → Heap
- Objekte allgemein → Heap
- Referenzen lokaler Objekte → Stack, Objekt selbst im Heap
- Lokale Variablen → Stack des ausführenden Threads (primitive Variablen in Java)
