## Funktionen sind First-Class Citizens
In Scala sind Funktionen *First-Class Citizens*. Das bedeutet:
- Sie können in Variablen gespeichert werden,
- als Eingabetyp in eine Funktion übergeben werden,
- als Rückgabetyp aus einer Funktion zurückgegeben werden.
=> Man kann mit ihnen umgehen wie mit Objekten.





````scala
//normale Funktion:
val f1: Int => Int = x => x + 1
val funf = 5
println(f1(3)) //Ausgabe: 4:Int, da 3+1
````


## Funktion höherer Ordnung
Eine Funktion höherer Ordnung ist eine Funktion, die **mindestens eine Funktion als Argument nimmt** oder **eine Funktion zurückgibt**.

### Geheimnis: Pfeile `=>` und Klammern `()` verstehen
- `=>` trennt:
  - **Eingabetyp vom Rückgabetyp**
  - **Eingabewert vom Rückgabewert**
- `()` gibt den Augen Halt

````scala
val multiply: (Int, Int) => Int = (x, y) => x * y
// Eingabetypen: (Int, Int)
// =>
// Rückgabetyp: Int
// -------
// Eingabewerte: (x, y)
// =>
// Rückgabewert: x * y

multiply(5,6)
````

## Funktion höherer Ordnung mit Funktion als Rückgabe:
- nimmt ein x vom Typ Int,
- gibt eine Funktion zurück, die sich dieses x merkt und noch auf y wartet

````scala
// Funktion höherer Ordnung:
val addSomeValueFunction: Int => (Int => Int) = 
    (x) =>  y => x + y
    
val fDreiGemerkt = addSomeValueFunction(3)
println(fDreiGemerkt(2))

val dreiGemerkt : Int => Int = y => y+3

println(fDreiGemerkt(2)) 

````

## Funktion höherer Ordnung mit Funktion als Eingabe
- Funktionen können auch weitere Funktionen als Eingabe bekommen.
- Diese Funktionen können dann innerhalb der Funktion genutzt werden.

````scala
val applyOperation: (Int, Int, ((Int, Int) => Int)) => Int =
  (x, y, funktion) => funktion(x, y)

applyOperation(2, 3, (x, y) => x + y) // 
applyOperation(2, 3, (x, y) => x * y) //
````

### Sekundärkonstruktor
Alternative Möglichkeit zur Objekterstellung mit this: muss in der ersten Zeile weiterhin den Primärkonstuktor aufrufen mit fehlenden Werten



````scala
class Animal(private val name: String,private val age: Int,private var species: String) {
    
    println("Ich bin noch Teil vom Primärkonstruktor")
    def this(name: String, age: Int) ={
        this(name, age,"unknown")
        println("Ich bin Teil vom Sekundärkonstruktor")
    }
    //Getter:
    def getAgePlusSomeValue(someValue:Int): Int = {age+someValue}
    //Methode:
    // Methode
    def methode: Unit =
        println(s"Ich bin eine Mehthode von $name dem $species.")

    def changeSpecies(newSpecies: String): Unit =
        species = newSpecies
}

val elvis = new Animal("Elvis", 5) // mittels new wird ein neues Objekt erstellt
elvis.getAgePlusSomeValue(5)
````

## Klassenvariablen

**Hinweis**: Companion Object hat Zugriff auf private Member der Klasse.

````scala
//Animal Companion Object
object Animal {
  var animalCount = 0
  private  def  countAnimal: Int = { // private so that only the class cna access it 
    animalCount += 1
    animalCount
  }
}
//in der Student Klasse:
class Animal(private val name: String, private val age: Int, private val species: String) {
  val id = Animal.countAnimal
} //companion object sees  private variables
    
val schurik = new Animal("Schurik",7,"Jack-Russell-Terrier")
val schurik2 = new Animal("Schurik",7,"Jack-Russell-Terrier")

println(schurik.id)

println(Animal.animalCount)


````

### Methoden/benannte Funktionen
- Eine benannte Funktion ist eine Funktion, die mit def definiert wird und einen festen Namen hat.
- Sie gehört zu einer Klasse bzw. zu einem Objekt und ist kein direktes „Funktions-Value“-Objekt wie eine Lambda.
- Sie können über ETA-Expansion dort genutzt werden, wo eigentlich LambdaFunktionen erwartet wären.

Die Deklaration und Initialisierung von benannten Funktionen funktioniert so:
```scala 
def nameDerFunktion(eingabewerte: Eingabetyp): Rückgabetyp = {
  Ausdruck
}
def nameDerFunktion: Eingabetypen => Rückgabetypen = (eingabewerte) => {Ausdruck}
```

````scala
object Rechner {

  // 🔹 Benannte Funktion (Methode)
    def verdoppeln(x: Int): Int = {
        x * 2
  }
  

    def halbieren(x:Int,y:Int):Int = {
        if (y==0) 0
        else x/y
        }
}

val halbieren: (Int,Int) => Int =  (x,y) => {
        if (y==0) 0
        else x/y
        }

List(1,2,3).map((x) => x*2 ) //higher order Function benötigt eigentlich normales Lambda 
List(1,2,3).map(Rechner.verdoppeln) // funktioniert wegen ETA-Expansion, wandelt Methode zu Funktion um
````
