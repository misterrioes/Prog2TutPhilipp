# Value Parametrization vs Behaviour Parametrization

- Bei Value Parametrization wird nur ein **Wert** übergeben. Die Logik bleibt in der Hauptfunktion. 
- Bei Behaviour Parametrization wird dagegen die Filterlogik selbst als **Strategie** übergeben.

In diesem Beispiel haben wir ein Array an Büchern und es wird innerhalb der Funktion selbst mit einem switch entschieden, welches Attribut geprüft wird.
```java
import java.util.LinkedList;

class Book {
    String author;
    String title;
    int year;

    public Book(String author, String title, int year) {
        this.author = author;
        this.title = title;
        this.year = year;
    }
}

public static LinkedList<Book> filterBooks(Book[] books, int type, Object value) {
    LinkedList<Book> result = new LinkedList<>();

    for (Book b : books) {
        switch (type) {
            case 1:
                if (b.author.equals(value)) result.add(b);
                break;
            case 2:
                if (b.title.equals(value)) result.add(b);
                break;
            case 3:
                if (b.year == (int) value) result.add(b);
                break;
        }
    }

    return result;
}
```

# Behaviour Parametrization
Bei der Behaviour Parametrization geben wir eine **Strategie** direkt mit in die Funktion. 

In Java geht das bei der klassischen Variante, indem wir uns eine Stragie/Funktionen definieren über zum Beispiel Interfaces, implementieren diese Interfaces unterschiedlich in verschiedenen Klassen und übergeben Objekte dieser Klassen an die Hauptfunktion. Diese Hauptfunktion weiß nur noch, dass diese Strategie existiert, aber die Implementierung ist in den Klassen gekapselt und nicht in der Hauptfunktion selbst.

```java
    interface BookFilterStrategy {
    boolean test(Book book);
} class AuthorFilter implements BookFilterStrategy {
    private String author;

    public AuthorFilter(String author) {
        this.author = author;
    }

    public boolean test(Book book) {
        return book.author.equals(author);
    }
}
class TitleFilter implements BookFilterStrategy {
    private String title;

    public TitleFilter(String title) {
        this.title = title;
    }

    public boolean test(Book book) {
        return book.title.equals(title);
    }
}
class YearFilter implements BookFilterStrategy {
    private int year;

    public YearFilter(int year) {
        this.year = year;
    }

    public boolean test(Book book) {
        return book.year == year;
    }
}
public static LinkedList<Book> filterBooks(Book[] books, BookFilterStrategy strategy) {
    LinkedList<Book> result = new LinkedList<>();

    for (Book b : books) {
        if (strategy.test(b)) {
            result.add(b);
        }
    }

    return result;
}
```

# Funktionale Interfaces und Lambda
## Funktionale Interfaces
Functional Interfaces in Java sind einfach Interfaces mit genau ***einer*** abstrakten Methode. 


Es ist möglich eigene Interfaces zu erstellen und zu nutzen Beispiel:  
```java


@functionalInterface
public interface BookFilterStrategy {
    boolean test(Book book); //eine abstrakte Methode => Interface = "funktionales Interface"
}


//Klasse TestOdd implementiert das F.I. und überschreibt die abstrakte Methode
public class YearAfterFilter implements BookFilterStrategy{
    private int year;
    public YearAfterFilter(int year) {
        this.year = year;
    }
    @Override
    public boolean test(Book book) {
        return book.getYear() > year;
    }
}
```

public class TestOdd implements Predicate <Integer> {

## Lambda Ausdrücke
Ein Lambda-Ausdruck ist eine anonyme Funktion, die ohne eigenen Namen definiert wird.

(Lambda-parameter) -> {Lambdarumpf}

## Lambda-Ausdruck in Java bedeutet immer auch Funktional-Interface
In Java kann eine anonyme Funktion nur verwendet werden um die einzige abstrakte Methode eines funktionalen Interfaces zu implementieren. Wenn der Kontext klar ist, bestimmt der Compiler, welches funktionale Interface gemeint ist.
```java
Predicate<Integer> isOdd = integer ->  (integer % 2) == 1; // Da Predicate ein funktionales Interface ist, kann der Lambda-Ausdruck automatisch der einzigen abstrakten Methode (test) zugeordnet werden.
System.out.println(isOdd.test(3));
```

#### Lambda-Ausdruck im Stream:
```java
int sum = Arrays.asList(1, 2, 3, 4, 5).stream()
    .map(n -> n * 2)         // Lambda gibt hier an, wie jedes Element verändert werden soll!
    .reduce(0, (a, b) -> a + b); // Lambda legt fest, wie summiert wird
System.out.println(sum); // Ausgabe: 30
```

# Funktionen in Scala
- Funktionen sind **First-Class Citizens**, das heißt sie können wie Werte behandelt, übergeben und zurückgegeben werden. 
- Im Gegensatz zu Java braucht man kein explizites funktionales Interface.
- Funktionen haben einen eigenen Typ, der durch die Typen ihrer Parameter und ihres Rückgabewerts bestimmt wird.


## Anonyme Funktionen
- Anonyme Funktionen sind namenlose Funktionsausdrücke, die als Werte existieren und direkt einer Variable zugewiesen oder als Argument übergeben werden können.
- Es gibt zwei Möglichkeiten anonyme Funktionen zu erstellen. Der Unterschied liegt nur darin, ob der Typ vorher festgelegt wird oder aus der Funktion selbst abgeleitet wird. 
```scala
val nameDerFunktion: Eingabetypen => Rückgabetypen = (eingabewerte) => {Ausdruck}
val nameDerFunktion = (eingabewert: Eingabetyp,...) => {Ausdruck}


````scala

val verdoppelnTypannotation: Int => Int = (x) => x*2 // Konkrete Typangabe 
val verdoppelnTypInferenz = (x:Int) => x*2// TypInferenz
val summeTypannotation: (Int, Int) => Int = (a,b) => a+b
val summeTypInferenz = (a: Int, b: Int) => a + b
val grussenTypann: String => Unit = (name)=> {
  println(s"Hallo $name!")}

verdoppelnBenannt(3) //Ausgabe: Int = 6
summeBenannt(2,2) //Ausgabe: Int = 4
grussenBenannt("Welt") //Ausgabe: Hallo Welt!
````

## anonyme Funktionen in Scala:

val nameDerFunktion = 
(eingabewerte: Eingabetyp) => {Ausdruck}

````scala
val nextInt: Int => Int = x => x + 1 // f ist vom Typ Lambda Int => Int
val multiplyString : (Int, String)  => String = (n, s) => {
    var result = ""
    for (_ <- 1 to n) {
      result += s
    } 
    result
  } // Rückgabetyp ist Lambda (Int,String)=> String

println(nextInt(100))
println(multiplyString(5, "Hallo Welt, "))
````

````scala
val isOdd = (x: Int) => x % 2 != 0
println(isOdd(3)) // Ausgabe: true

val list = List(1, 2, 3, 4, 5)
println(list.filter(isOdd)) // filter Methode erwartet eine Funktion vom Typ der in der Liste ist zu Boolean
println(list.filter((x: Int) => x % 2 != 0)) //dasselbe nur ohne Zwischenspeichern der Funktion

val sum = List(1, 2, 3, 4, 5)
  .map(n => n * 2)       // Lambda: wie jedes Element verändert wird
  .reduce((a, b) => a + b) // Lambda: wie kombiniert wird

println(sum) // Ausgabe: 30
````
